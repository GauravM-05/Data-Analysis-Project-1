```python
# 1. Find the number of distinct bookings, sessions, and searches from the given data sets.

distinct_bookings = bookings.booking_id.nunique()
sessionss = sessions.session_id.nunique()
searches = sessions.search_id.nunique()

print("No. of distinct bookings:", distinct_bookings)
print("No. of sessions:", sessionss)
print("No. of searches:", searches)
```

```python
# 2. How many sessions have more than one booking?

bs_merge = sessions.merge(bookings, on = "booking_id", how = "left")
bs_merge
```

```python
booking_sessions = bs_merge.groupby("booking_id").agg(Bookings_per_sessions = ("session_id", "count"))
booking_sessions.value_counts()
```

```python
print("The no. of sessions more than one bookings are:", booking_sessions[booking_sessions>1].count())
```

```python
# 9. Plot the trends at a quarterly frequency for the number of bookings by each of the
# device types, that is, plot a time series for each year and quarter showing the
# number of bookings performed by each device type.

bookings.set_index('booking_time_new', inplace=True)

quarterly_device_bookings = bookings.groupby('device_type_used').resample('Q').size().unstack(fill_value=0)

quarterly_device_bookings
```

```python
plt.figure(figsize=(14, 18))
quarterly_device_bookings.T.plot(kind='line')
plt.title('Quarterly Number of Bookings by Device Type')
plt.xlabel('Date')
plt.ylabel('Number of Bookings')
plt.legend(title='Device Type')
plt.show()
```

```python
# 10. Consider the following example:

# 12 customers performed a total of 100 searches but only a few of them performed
# 10 bookings in total.
# Searches = 100 ,bookings = 10 ,customers = 12
# The overall booking to search ratio, Ober, is 10/100 = 0.1 Using the above
# information/context, answer the following:
# • What is the average oBSR for each month of the year?
# • What is the average oBSR for each day of the week?
# • Plot a time series of oBSR on all the given dates.

booking_session_merge = sessions.merge(bookings, on = 'booking_id', how = 'left')
booking_session_merge
```

```python
booking_session_merge.info()
```

```python
booking_session_merge['search_time'] = pd.to_datetime(booking_session_merge['search_time'], errors='coerce')
booking_session_merge['booking_time'] = pd.to_datetime(booking_session_merge['booking_time'], errors='coerce')

searches = booking_session_merge.groupby(booking_session_merge['search_time'].dt.date).size()
searches
```

```python
daily_bookings =booking_session_merge.groupby(booking_session_merge['booking_time'].dt.date).size()
print("The daily bookings are : \n",daily_bookings)
```

```python
daily_oBSR = daily_bookings/searches
print("Daily oBSR are : \n\n",daily_oBSR)
```

```python
daily_oBSR.index = pd.to_datetime(daily_oBSR.index)
monthly_oBSR = daily_oBSR.resample('M').mean()
print("Monthly oBSR are : \n\n",monthly_oBSR)
```

```python
day_oBSR = daily_oBSR.groupby(daily_oBSR.index.day_name()).mean()
print("Day oBSR are : \n\n",day_oBSR)
```

```python
plt.figure(figsize=(14,8))
daily_oBSR.plot(kind = 'line')
plt.title('Time Series of Overall Booking to Search Ratio (oBSR)')
plt.xlabel('Date')
plt.ylabel('oBSR')
plt.grid(True)
plt.show()
```
